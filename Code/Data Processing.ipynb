# EDA & Preprocessing

```python
import numpy as np
import pandas as pd
```

# Read Q1 Data


```python
df1 = pd.read_csv("q1.csv")
df1
```

```python
selected_stations = [68,78,83,85,88,97]
```

```python
df1_f = df1[df1["STATION ID"].isin(selected_stations)]
```

# Read Q2 Data


```python
df2 = pd.read_csv("q2.csv")
df2_f = df2[df2["STATION ID"].isin(selected_stations)]
df2_f
```

# Merging both Q1 and Q2

```python
df = pd.concat([df1_f,df2_f])
```

# Resampling for 1Hr

```python
usage = df[["STATION ID","TIME","AVAILABLE BIKE STANDS"]]
usage['TIME'] = pd.to_datetime(usage['TIME'],format='%Y-%m-%d %H:%M:%S')
usage.index = usage['TIME']
```

```python
usage_value = pd.DataFrame()

for id in usage["STATION ID"].unique():
    
    usage_d = usage[usage["STATION ID"] == id]
    tmp = usage_d.resample("1H").mean()
    
    if usage_value.empty:
        usage_value = tmp
    else:
        usage_value = pd.concat([usage_value,tmp])
    
```

```python
usage_value.dropna(inplace = True)
usage_value
```

# Weather Data

```python
w_data = pd.read_csv("weatherData.csv")
```

```python
w_data['date'] = pd.to_datetime(w_data['date'])
```

```python
w_data.index = w_data['date']
```

```python
## Merging weather and usage data
data = usage_value.merge(w_data,left_index = True,right_index=True)
```

```python
data
```

# OneHotEncoding for Station ID

```python
df_s = df['STATION ID'].unique()
stations = pd.DataFrame(df_s)
y = pd.get_dummies(stations[0], prefix="STATION")
s_code = pd.merge(left=stations,right=y,left_index=True,right_index = True)
s_code = s_code.rename(columns={0:"STATION ID"})
data = pd.merge(left=data,right=s_code)
```

```python
data
```

# Generating Features W1-W5

```python
n_df = data
```

```python
n_df.rename(columns = {'AVAILABLE BIKE STANDS':'USAGE'}, inplace = True)
n_df['date'] = pd.to_datetime(n_df['date'],format='%d-%m-%Y %H:%M')
n_df.index = n_df['date']
```

```python
def getYValue(station_id,date,n):
    value = None
    new_date = date - pd.to_timedelta(n,unit='d')
    data = n_df[(n_df['STATION ID'] == station_id)& (n_df['date'] == new_date)]
    if data.empty:
        pass
    else:
        value = data['USAGE'][0]
    
    return value
    
```

```python
n_df['W1'] = n_df.apply( lambda x: getYValue(x['STATION ID'],x['date'],1*7), axis = 1)
```

```python
n_df['W2'] = n_df.apply( lambda x: getYValue(x['STATION ID'],x['date'],2*7), axis = 1)
n_df['W3'] = n_df.apply( lambda x: getYValue(x['STATION ID'],x['date'],3*7), axis = 1)
n_df['W4'] = n_df.apply( lambda x: getYValue(x['STATION ID'],x['date'],4*7), axis = 1)
n_df['W5'] = n_df.apply( lambda x: getYValue(x['STATION ID'],x['date'],5*7), axis = 1)
```
